```python
df = pd.read_json(r'E:\Learn\DS4200\GroupPorject\yelp_academic_dataset_business.json', lines=True)
df = df[df['categories'].str.contains('Restaurants', na=False)].dropna()
```

```python
df = df.drop(columns = [ 'business_id']).reset_index(drop=True)
```

```python
#filters out all attribute values with False
def attribute_filter(attributes):
    for k in list(attributes):
        if attributes[k] == 'False' or attributes[k] =='None' or attributes[k] == "u'none'":
             attributes.pop(k, None)
        elif attributes[k][0] == '{':
             attributes.pop(k, None)
    return attributes
```

```python
def hours_filter(hours):
    for k in list(hours):
        if hours[k] == '0:0-0:0':
            hours.pop(k, None)
    if bool(hours):
        return hours
    else:
        return 
```

```python
def categories_conversion(categories):
    categories = categories.replace(" ", "").split(",")
    return categories
    
```

```python
df['attributes'] = df['attributes'].apply(lambda x: attribute_filter(x))
df['hours'] = df['hours'].apply(lambda x: hours_filter(x))
df['categories'] = df['categories'].apply(lambda x: categories_conversion(x))
```

```python
df = df.dropna().reset_index(drop=True)
```
